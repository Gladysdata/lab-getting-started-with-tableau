```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn  as sns
```

 - Load the dataset into a notebook

```python
df = pd.read_csv('we_fn_use_c_marketing_customer_value_analysis.csv')
df.head()
```

 - Create a barplot of the number of customers per Gender

```python
sns.countplot(data=df,x='Gender')
 
plt.title("Number of customers per Gender")
plt.show()
```

 - Create a barplot of the number of customers per EmploymentStatus and Gender

```python
sns.histplot(data=df, x="Gender", color="red")
sns.histplot(data=df, x="EmploymentStatus", color="green")
plt.title("Number of customers per Employment Status and Gender")
plt.show()
```
